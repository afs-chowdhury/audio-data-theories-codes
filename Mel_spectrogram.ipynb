```python
# importing libraries

import torch
import torchaudio
import os
import numpy as np
import librosa
import matplotlib.pyplot as plt
import librosa.display
import IPython.display as ipd
```

```python
DATA_DIR = "audio"
SCALE_FILE = "scale.wav"
REDHOT_FILE = "redhot.wav"
VOICE_FILE = "voice.wav"
DEBUSSY_FILE =  "debussy.wav"
DUKE_FILE = "duke.wav"
```

```python
scale , sr = librosa.load(os.path.join(DATA_DIR,SCALE_FILE))
```

```python
FRAME_SIZE = 2048
FILTER_BANK = 10
SR = sr
HOP_LENGTH = 512
```

### Mel Filete Banks

```python
filter_banks = librosa.filters.mel(sr = SR,
                                  n_fft=FRAME_SIZE,
                                  n_mels=FILTER_BANK)
```

```python
filter_banks.shape
```

```python
plt.figure(figsize = (25,10))
librosa.display.specshow(filter_banks,
                        x_axis= 'linear')
plt.colorbar();
```

### Extracting Mel Spectrogram

```python
mel_spec = librosa.feature.melspectrogram(scale,sr =sr,n_fft=FRAME_SIZE,hop_length=HOP_LENGTH)
```

```python
mel_spec.shape
```

if the filter bank number is not given (n_mels) then it will make total 128 filter bank 

```python
log_mel_spec = librosa.amplitude_to_db(mel_spec)
plt.figure(figsize = (25,10))
librosa.display.specshow(log_mel_spec,
                        x_axis= 'time',
                        y_axis='mel')
plt.colorbar();
```

```python




def plot_spectrogram_librosa(spec):
    spec = spec.numpy()
    spec_db = librosa.amplitude_to_db(spec)
    plt.figure(figsize=(12, 4))
    librosa.display.specshow(spec_db, x_axis='time', y_axis='log')
    plt.colorbar(format='%+2.0f dB');
    plt.title('STFT Spectrogram')
    plt.tight_layout()
    plt.show()
```
